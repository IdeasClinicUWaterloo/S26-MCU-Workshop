## Introducing the ALU

We wish to build a simple "ALU" (arithmetic logic unit).
In digital design we always start our design from a black box with inputs and outputs.
This is the **interface** and designing it carefully is one of the key steps to making sure what we build is useful.
Defining how the inputs and outputs interact over time with other components is done using a **protocol** (specification of how changes are handled) and is also important.

In this exercise the interface and protocol are both already specified.
Your job will be writing the implementation of the interface, making sure that it _behaves_ according to the the protocol we give you.

This is the interface for the ALU.

A functional description of the ALU might be:

> The ALU takes in binary numbers `a` and `b`. It then performs an operation: it may:
>
> 0. Add `a` and `b`. (`a + b`)
> 1. Subtract `b` from `a`. (`a - b`)
> 2. Multiply `a` and `b`. (`a * b`)
> 3. Right shift `a` by `b`. (`a >> b`)
>
> Which of these operations is performed is selected by using the `op_sel` input.
>
> - If the binary number given to `op_sel` is the same as the decimal number 0, add.
> - If the binary number given to `op_sel` is the same as the decimal number 1, subtract.
>
> And so on.
